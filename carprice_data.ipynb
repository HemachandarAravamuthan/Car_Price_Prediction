```python
# Refined columns
new_det.head()
```

```python
# Function to convert string price column to int
def conversion(x):
  if x.endswith(' Crore'):
    return(int((float(x.replace('₹','').replace('Crore','').replace(',','')))*10000000))
  elif x.endswith(' Lakh'):
    return(int((float(x.replace('₹','').replace(' Lakh','').replace(',','')))*100000))
  elif x.endswith(' Thousand'):
    return(int((float(x.replace('₹','').replace(' Thousand','').replace(',','')))*1000))
```

```python
# Applyinng the funtion to the price column
new_det['price']=new_det['price'].apply(conversion)
```

```python
# Function to convert km in int
def distance(x):
  return(int(x.replace(',','')))
```

```python
# Applying the function
new_det['km']=new_det['km'].apply(distance)
```

### Feature Engineering

```python
# Datatypes
dff.dtypes
```

```python
# Null values
dff.isna().sum()
```

```python
# Fueltype
dff['fuelType'].value_counts()
```

```python
# Importing preprocessing
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
# Coding fueltype column
ft_code=le.fit_transform(dff['fuelType'])
dff.insert(loc=1,
          column='ft_code',
          value=ft_code)
```

```python
# Bodytype
dff['bodyType'].value_counts()
```

```python
# Encoding bodytype column
bt_code=le.fit_transform(dff['bodyType'])
dff.insert(loc=3,
          column='bt_code',
          value=bt_code)
```

```python
# Transmission
dff['transmission'].value_counts()
```

```python
# Encoding transmission column
t_code=le.fit_transform(dff['transmission'])
dff.insert(loc=6,
          column='t_code',
          value=t_code)
```

```python
# Manufacturer
dff['manufacturer'].value_counts()
```

```python
# Encoding manufacturer column
m_code=le.fit_transform(dff['manufacturer'])
dff.insert(loc=9,
          column='m_code',
          value=m_code)
```

```python
# City
dff['city'].value_counts()
```

```python
# Encoding city column
c_code=le.fit_transform(dff['city'])
dff.insert(loc=15,
          column='c_code',
          value=c_code)
```

```python
# Dataframe
dff.head()
```

```python
# Correlation heatmap
corr_df=dff[['ft_code','bt_code','km','t_code','ownerNo','m_code','modelYear','centralVariantId','c_code','price']].dropna().corr()
f, ax = plt.subplots(figsize=(10, 7))
sns.heatmap(corr_df, annot=True, linewidths=.5, ax=ax)
```

```python
dff=dff.dropna()
```

```python
# Assigning the values
X=dff[['ft_code','bt_code','km','t_code','ownerNo','m_code','modelYear','centralVariantId','c_code']].dropna()
y=dff['price']
```

### Model building

```python
# Scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit_transform(X)
```

```python
# Splitting the data
# Importing the split
from sklearn.model_selection import train_test_split

# Objecting the split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 4)
```

```python
# Create a regressor
from sklearn.linear_model import LinearRegression

# Create a regressor object
lin_reg = LinearRegression()

# fit the model with X-train and y_train data
lin_reg.fit(X_train,y_train)
```

```python
# import the module
from sklearn.metrics import mean_absolute_error as mae

pred=lin_reg.predict(X_test)
error=mae(y_test,pred)
print(error)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg=DecisionTreeRegressor()

dt_reg.fit(X_train,y_train)
```

```python
pred1=dt_reg.predict(X_test)
error=mae(y_test,pred1)
print(error)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg=RandomForestRegressor(n_estimators=100, random_state=42)

rf_reg.fit(X_train,y_train)
```

```python
pred2=rf_reg.predict(X_test)
error=mae(y_test,pred2)
print(error)
```

```python
# Importing the model
from sklearn.ensemble import GradientBoostingRegressor
# Objecting the model
gbr = GradientBoostingRegressor(n_estimators = 10, learning_rate = 0.1, random_state =42)
# Train with dataset
gbr.fit(X_train,y_train)
```

```python
pred3=gbr.predict(X_test)
error=mae(y_test,pred3)
print(error)
```
